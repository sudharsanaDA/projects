```python
amt=10000
pin=1234
name=('sana')
ac=('xxxx xxxx 8256')
print("welcome to IOB ATM")
opt=int(input("Select option\n1.Withdraw   2.Deposit\n3.Balance  4.Cancel" ))
if opt==1:
  u_amt=int(input("Enter Amount:"))
  u_pin=int(input("Enter Pin Number"))
  if pin==u_pin:
    if amt>=u_amt:
      print("Withdraw Success")
      r=input("Do you need to see the receipt?")
      if r=="yes":
        print("Holder Name:",name)
        print("Acc.No:",ac)
        print("Old Balance:",amt)
        print("New Balance:",amt-u_amt)
      else:
        print("Thank You")
    else:
      print("Insufficient Balance")
  else:
    print("Invalid Pin")
elif opt==2:
  u_amt=int(input("Enter Amount:"))
  u_pin=int(input("Enter Pin Number"))
  if pin==u_pin:
    print("Deposit Success")
    r=input("Do you need to see the receipt?")
    if r=="yes":
        print("Holder Name:",name)
        print("Acc.No:",ac)
        print("Old Balance:",amt)
        print("New Balance:",amt+u_amt)
    else:
        print("Thank You")
  else:
    print("Invalid Pin")
elif opt==3:
  u_pin=int(input("Enter Pin Number"))
  if pin==u_pin:
    print("Balance:",amt)
  else:
    print("Invalid Pin")
elif opt==4:
  print("Thank You")
```
