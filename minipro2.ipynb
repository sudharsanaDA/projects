```python
def fruit_shop():

    fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]
    prices = [120, 40, 150, 60, 90]  # prices per kg


    cart = []

    while True:
        print("\nAvailable Fruits:")
        for idx, fruit in enumerate(fruits):
            print(f"{idx + 1}. {fruit} - ₹{prices[idx]}/kg")


        try:
            choice = int(input("Enter the number of the fruit you want to buy: ")) - 1
            if choice < 0 or choice >= len(fruits):
                print("Invalid choice. Please try again.")
                continue
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue


        try:
            qty = float(input(f"How many kgs of {fruits[choice]} do you want? "))
            if qty <= 0:
                print("Quantity must be positive.")
                continue
        except ValueError:
            print("Invalid quantity.")
            continue


        item_total = prices[choice] * qty
        cart.append({
            "fruit": fruits[choice],
            "price": prices[choice],
            "quantity": qty,
            "total": item_total
        })
        print(f"Added {qty} kg of {fruits[choice]} - ₹{item_total:.2f}")


        more = input("Do you want to buy anything else? (yes/no): ").strip().lower()
        if more != 'yes':
            break


    print("\n--- Final Bill ---")
    grand_total = 0
    for item in cart:
        print(f"{item['fruit']} - {item['quantity']} kg x ₹{item['price']} = ₹{item['total']:.2f}")
        grand_total += item['total']
    print(f"Total Amount: ₹{grand_total:.2f}")
    print("Thank you for shopping with us")

fruit_shop()
```

```python
# prompt: Do the above function without using f function

def fruit_shop():

    fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]
    prices = [120, 40, 150, 60, 90]  # prices per kg


    cart = []

    while True:
        print("\nAvailable Fruits:")
        for idx, fruit in enumerate(fruits):
            print("{}. {} - ₹{}/kg".format(idx + 1, fruit, prices[idx]))


        try:
            choice = int(input("Enter the number of the fruit you want to buy: ")) - 1
            if choice < 0 or choice >= len(fruits):
                print("In valid choice. Please try again.")
                continue
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue


        try:
            qty = float(input("How many kgs of {} do you want? ".format(fruits[choice])))
            if qty <= 0:
                print("Quantity must be positive.")
                continue
        except ValueError:
            print("Invalid quantity.")
            continue


        item_total = prices[choice] * qty
        cart.append({
            "fruit": fruits[choice],
            "price": prices[choice],
            "quantity": qty,
            "total": item_total
        })
        print("Added {} kg of {} - ₹{:.2f}".format(qty, fruits[choice], item_total))


        more = input("Do you want to buy anything else? (yes/no): ").strip().lower()
        if more != 'yes':
            break


    print("\n--- Final Bill ---")
    grand_total = 0
    for item in cart:
        print("{} - {} kg x ₹{} = ₹{:.2f}".format(item['fruit'], item['quantity'], item['price'], item['total']))
        grand_total += item['total']
    print("Total Amount: ₹{:.2f}".format(grand_total))
    print("Thank you for shopping with us")

fruit_shop()
```

```python
# Movie class
class Movie:
    def __init__(self, title, time, price):
        self.title = title
        self.time = time
        self.price = price

    def show_details(self, index):
        print(str(index) + ". " + self.title + " at " + self.time + " - ₹" + str(self.price) + " per ticket")


# Screen class with multiple movies
class Screen:
    def __init__(self, name, movies):
        self.name = name
        self.movies = movies  # List of Movie objects

    def show_movies(self):
        print("\n" + self.name + " Shows:")
        for i in range(len(self.movies)):
            self.movies[i].show_details(i + 1)


# Billing class
class Billing:
    def __init__(self, screen, movie, tickets):
        self.screen = screen
        self.movie = movie
        self.tickets = tickets

    def generate_bill(self):
        total = self.tickets * self.movie.price
        print("\n--- Ticket Bill ---")
        print("Screen: " + self.screen.name)
        print("Movie: " + self.movie.title)
        print("Show Time: " + self.movie.time)
        print("Tickets: " + str(self.tickets))
        print("Price per ticket: ₹" + str(self.movie.price))
        print("Total Amount: ₹" + str(total))
        print("Enjoy your movie!")


# Create movies for screen 1
movies_screen1 = [
    Movie("Inception", "6:00 PM", 150),
    Movie("Avatar", "8:30 PM", 170),
    Movie("Iron Man", "10:30 PM", 140)
]

# Create movies for screen 2
movies_screen2 = [
    Movie("Interstellar", "5:00 PM", 160),
    Movie("Dune", "7:45 PM", 180),
    Movie("The Batman", "10:15 PM", 200)
]

# Create screen objects
screen1 = Screen("Screen 1", movies_screen1)
screen2 = Screen("Screen 2", movies_screen2)

# Show screens and movies
print("Welcome to the Movie Booking System!")
print("\nAvailable Screens:")
print("1. " + screen1.name)
print("2. " + screen2.name)

# Screen selection
screen_choice = input("Choose a screen (1 or 2): ")

if screen_choice == "1":
    selected_screen = screen1
elif screen_choice == "2":
    selected_screen = screen2
else:
    print("Invalid screen selection.")
    exit()

# Show movies on the selected screen
selected_screen.show_movies()
movie_choice = int(input("Select the movie number: ")) - 1

# Validate movie choice
if movie_choice < 0 or movie_choice >= len(selected_screen.movies):
    print("Invalid movie selection.")
    exit()

selected_movie = selected_screen.movies[movie_choice]

# Ticket count
tickets = int(input("How many tickets do you want?: "))

# Generate bill
bill = Billing(selected_screen, selected_movie, tickets)
bill.generate_bill()
```
