```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("students_exam_results.csv")
print(data)
```

```python
data.fillna(35, inplace=True)
```

```python
data['Total'] = data[['Math', 'Science', 'English']].sum(axis=1)
print(data['Total'])
```

```python
print(data)
```

```python
highest=data.sort_values(by='Total', ascending=False).head(1)
print(highest)
```

```python
plt.bar(data['Student_ID'],data['Total'],color="green")
plt.title("Students performance ")
plt.xlabel("Student Id")
plt.ylabel("Marks")
plt.show()
```
