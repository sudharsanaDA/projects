```python
# prompt: Give code for the above same concept Some better code
class User:
    """Represents a user in the system."""
    def __init__(self, name, pickup, drop):
        self.name = name
        self.pickup = pickup
        self.drop = drop

class Ride:
    """Represents a ride with distance and bill calculation."""
    def __init__(self, user, distance):
        self.user = user
        self.distance = distance
        self._rate_per_km = 10  # Use a protected attribute for rate

    def calculate_bill(self):
        """Calculates the total bill for the ride."""
        return self.distance * self._rate_per_km

    def display_bill(self):
        """Displays the ride bill details."""
        total_bill = self.calculate_bill()

        print(f"\n----- Ride Bill -----")
        print(f"Name            : {self.user.name}")
        print(f"From            : {self.user.pickup}")
        print(f"To              : {self.user.drop}")
        print(f"Distance        : {self.distance} km")
        print(f"Rate per KM     : ₹{self._rate_per_km}")
        print(f"Total Amount    : ₹{total_bill}")
        print("Thank you for using our service!")

# --- Main Execution ---
if __name__ == "__main__":
    # Get user input
    user_name = input("Enter your name: ")
    pickup_location = input("Enter pickup location: ")
    drop_location = input("Enter drop location: ")

    while True:
        try:
            ride_distance = float(input("Enter distance in KM: "))
            if ride_distance < 0:
                print("Distance cannot be negative. Please enter a valid distance.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter a numerical value for distance.")

    # Create objects and process the ride
    user_info = User(user_name, pickup_location, drop_location)
    current_ride = Ride(user_info, ride_distance)
    current_ride.display_bill()
```
