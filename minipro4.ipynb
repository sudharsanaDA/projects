```python

class Cart(list):
    def __init__(self):
        super().__init__()
        self.total = 0
    def add(self, item, qty, price):
        self.append((item, qty, qty * price))
        self.total += qty * price
    def bill(self):
        for i, q, c in self: print(f"{q} x {i} = ${c}")
        print("Total:", self.total)

class Supermarket:
    def __init__(self): self.cart = Cart()
    def buy(self, items):
        while True:
            print("\nItems:", ', '.join(f"{k}(${v})" for k,v in items.items()))
            i = input("Item: ").strip()
            q = input("Qty: ")
            if i in items and q.isdigit() and int(q) > 0:
                self.cart.add(i, int(q), items[i])
            else:
                print("Invalid.")
                continue
            if input("More? (yes/no): ") != 'yes': break

class Floor1(Supermarket): items = {'Apple': 3, 'Banana': 2}
class Floor2(Supermarket): items = {'Rice': 1, 'Bread': 2}
class Floor3(Supermarket): items = {'Soap': 2, 'Shampoo': 5}
class Floor4(Supermarket): items = {'Pen': 1, 'Notebook': 4}
floors = {1: Floor1, 2: Floor2, 3: Floor3, 4: Floor4}

def main():
    s = Supermarket()
    while True:
        f = input("\nChoose floor (1-4): ")
        if f.isdigit() and int(f) in floors:
            fl = floors[int(f)]()
            fl.cart = s.cart
            fl.buy(fl.items)
        else:
            print("Invalid floor.")
        if input("Another floor? (yes/no): ") != 'yes': break
    s.cart.bill()

main()
```
